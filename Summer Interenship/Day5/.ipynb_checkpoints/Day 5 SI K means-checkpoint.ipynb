```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('C:\\Users\\dell\\Downloads\\Mall_Customers.csv')
x = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss =[]
for i in range(1,11):
    kmeans=KMeans(n_clusters = i, init ='k-means++',random_state =10)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#fitting k-means to the dataset
kmeans =KMeans(n_clusters = 5, init ='k-means++', random_state =42)
y_kmeans = kmeans.fit_predict(x)

#visualisation the clusters
plt.scatter(x[y_kmeans ==0, 0], x[y_kmeans ==0, 1], s = 100, c= 'red', label = 'Faithful customres')
plt.scatter(x[y_kmeans ==1, 0], x[y_kmeans ==1, 1], s = 100, c= 'blue', label = 'Targets')
plt.scatter(x[y_kmeans ==2, 0], x[y_kmeans ==2, 1], s = 100, c= 'green', label = 'smart ppl')
plt.scatter(x[y_kmeans ==3, 0], x[y_kmeans ==3, 1], s = 100, c= 'cyan', label = 'happy cust')
plt.scatter(x[y_kmeans ==4, 0], x[y_kmeans ==4, 1], s = 100, c= 'magenta', label = 'premium clients')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s =300, c = 'yellow', label = 'Centroids')

plt.title('cluster of customers')
plt.xlabel('annual income(k$)')
plt.ylabel('spending score (1-100)')
plt.legend()
plt.show()
```

# Decision tree iris 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

color =['sepal_length', 'sepal_width', 'petal_length','petal_width', 'flower']
df = pd.read_csv('C:\\Users\\dell\\Downloads\\iris.data.txt', header=None, names=color)
print(df['flower'].unique())

flower_mapping = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
df['flower'] =df['flower'].apply(lambda x: flower_mapping[x])
df
```

```python
iris_setosa = df[df['flower']==0]
iris_virginica = df[df['flower']==1]
iris_versicolor =df[df['flower']==2]
print(len(iris_setosa))
print(len(iris_virginica))
print(len(iris_versicolor))

plt.scatter(iris_setosa['petal_length'], iris_setosa['petal_width'], color='r', label='Setosa')
plt.scatter(iris_virginica['petal_length'], iris_virginica['petal_width'] ,color='g', label='virginica')
plt.scatter(iris_versicolor['petal_length'], iris_versicolor['petal_width'] ,color='b', label='versicolor')
plt.legend()
plt.grid()
plt.xlabel('Petal Length')
plt.ylabel('PetalWidth')
```

```python
x_train, x_test,y_train, y_test = train_test_split(df[['petal_length', 'petal_width']],df['flower'],test_size=0.2)
y_train.value_counts().plot(kind='bar')
print(x_test)

decisiontree = DecisionTreeClassifier(criterion='entropy',max_depth=2)
obtained_tree = decisiontree.fit(x_train,y_train)
```

```python
from sklearn.tree import export_graphviz
from io import StringIO
import pydotplus
from IPython.display import Image

DecisionTree = StringIO()

export_graphviz(obtained_tree, out_file=DecisionTree,
                filled=True, rounded=True , feature_names=color[2:-1],
               special_characters=True)
graph = pydotplus.graph_from_dot_data(DecisionTree.getvalue())
Image(graph.create_png())
```
