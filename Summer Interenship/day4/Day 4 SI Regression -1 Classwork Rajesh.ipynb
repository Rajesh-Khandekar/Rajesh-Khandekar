```python
import matplotlib.pyplot as plt

PSize=[[6],[8],[10],[14],[18]]
PPrice= [7,9,13,17.5,18]

plt.figure()
plt.title('Pizza prize plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollers')
plt.plot(PSize, PPrice,'go')
plt.axis([0,25,0,25])
plt.grid(True)
```

```python
from sklearn.model_selection import train_test_split
PSize_train, PSize_test, PPrice_train, PPrice_test = \
train_test_split(PSize, PPrice, test_size=0.2)

plt.figure()
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.plot(PSize_train, PPrice_train, 'go')
plt.axis([0,25,0,25])
plt.grid(True)
PSize_test
```

```python
from sklearn.linear_model import LinearRegression

#create and fit the model
model = LinearRegression()
model.fit(PSize_train,PPrice_train)

print(PPrice_test)                                     ##test output = exoected output
model.predict(PSize_test)                              ##observed output = its a single list
```

```python
# plot initail data
plt.figure()
plt.title('Pizza Price prediction against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollers')
plt.plot(PSize_train, PPrice_train, 'go')
plt.plot(PSize_test, PPrice_test, 'bo')

plt.axis([0,25,0,25])
plt.grid()

#plot regression solution line
regression_line_y = model.predict([[0],[25]])
#print(regression_line_y)
plt.plot([0,25],regression_line_y,
         color='r',linestyle='-',linewidth=1)
```

```python
b= 16.75
a= model.predict([[b]])
print('A 16 pizza should cost: $%.2f' % a)   #prediction of inexperienced data
```

```python
model.score(PSize,PPrice)
```

```python
print("Slope(m): %.2f, Intercept (c):%.2f"%
     (model.coef_, model.intercept_))
```

```python
#simple linear regression

#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# importing the dataset
dataset = pd.read_csv('C:\\Users\\dell\\Downloads\\Salary_Data.csv')
x= dataset.iloc[:, :-1].values
y= dataset.iloc[:,1].values
print(x)
print(y)
```

```python
# splitting the dataset into the training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 1/3, random_state=0)

#fitting simple linear regression to the training set
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train, y_train)

#predicting the test set result
y_pred = regression.predict(x_test)
```

```python
# visualising the test set result
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regression.predict(x_train),color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```
